15.Выполните задание с применением собственных функций на основе
набора данных films.csv:
Кто из режиссеров снял больше всего фильмов?

```python
import pandas as pd

df = pd.read_csv("/home/films.csv")

director_counts = df['director'].value_counts()
director_counts
```

```python
most_films = director_counts.max()
top_directors = director_counts[director_counts == most_films]

print("Режиссёр(ы), снявший(е) больше всего фильмов:\n", top_directors)
```

```python
films_data = pd.read_csv('/home/films.csv')

films_data['director'] = films_data['director'].str.split('|')

exploded_directors = films_data.explode('director')

director_counts = exploded_directors['director'].value_counts()

most_prolific_director = director_counts.idxmax()
number_of_films = director_counts.max()

print(f"Режиссер, снявший больше всего фильмов: {most_prolific_director}")
print(f"Количество фильмов: {number_of_films}")
```
